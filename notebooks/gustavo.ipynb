```python
# Re-execute the plotting helper after environment reset.
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

def plot_3d_from_df(
    data: pd.DataFrame,
    parallax_unit: str = "mas",
    min_parallax: float = 0.0,
    max_distance_pc: float | None = None,
    clip_percentile: float = 99.5,
    sample_max: int | None = 200_000,
):
    """
    Make a physically meaningful 3D plot from a catalog with ra, dec, parallax.

    Parameters
    ----------
    data : DataFrame
        Must contain columns: 'ra' [deg], 'dec' [deg], 'parallax' [mas or arcsec].
    parallax_unit : {"mas","arcsec"}
        Unit of the parallax column.
    min_parallax : float
        Minimum parallax to keep (in the provided unit). Non-positive parallaxes are dropped by default.
    max_distance_pc : float | None
        If provided, cap distances at this value (pc). Useful to remove far outliers.
    clip_percentile : float
        If max_distance_pc is None, distances will be clipped to this percentile to remove extreme tails.
    sample_max : int | None
        If the dataset is huge, take a random sample up to this many rows for plotting.
    """
    df = data.copy()

    # Basic column checks
    required = {"ra", "dec", "gaia_parallax"}
    missing = required - set(df.columns)
    if missing:
        raise ValueError(f"Missing required columns: {sorted(missing)}")

    # Drop non-finite values
    df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=["ra", "dec", "gaia_parallax"])

    # Enforce positive parallax (or a minimum threshold)
    if min_parallax is None:
        min_parallax = 0.0
    df = df[df["gaia_parallax"] > float(min_parallax)]

    if df.empty:
        raise ValueError("No rows left after filtering invalid/negative parallaxes.")

    # Distance in parsec
    if parallax_unit.lower() == "mas":
        df["distance_pc"] = 1000.0 / df["gaia_parallax"].astype(float)
    elif parallax_unit.lower() == "arcsec":
        df["distance_pc"] = 1.0 / df["gaia_parallax"].astype(float)
    else:
        raise ValueError("parallax_unit must be 'mas' or 'arcsec'.")

    # Remove extreme distances
    if max_distance_pc is not None:
        df = df[df["distance_pc"] <= float(max_distance_pc)]
    else:
        dcap = np.nanpercentile(df["distance_pc"], clip_percentile)
        df = df[df["distance_pc"] <= dcap]

    if df.empty:
        raise ValueError("No rows left after distance clipping. Consider raising max_distance_pc or percentile.")

    if sample_max is not None and len(df) > sample_max:
        df = df.sample(sample_max, random_state=42)

    ra_rad = np.deg2rad(df["ra"].astype(float).to_numpy())
    dec_rad = np.deg2rad(df["dec"].astype(float).to_numpy())
    d_pc = df["distance_pc"].to_numpy()

    x = d_pc * np.cos(dec_rad) * np.cos(ra_rad)
    y = d_pc * np.cos(dec_rad) * np.sin(ra_rad)
    z = d_pc * np.sin(dec_rad)

    fig = plt.figure(figsize=(8, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z, s=1, alpha=0.5)

    ax.set_xlabel("x [pc]")
    ax.set_ylabel("y [pc]")
    ax.set_zlabel("z [pc]")
    ax.set_title("3D positions from (ra, dec, parallax)")

    max_range = np.array([x.max()-x.min(), y.max()-y.min(), z.max()-z.min()]).max()
    Xb = 0.5*max_range
    for axis, center in zip((ax.set_xlim, ax.set_ylim, ax.set_zlim),
                            ((x.max()+x.min())/2, (y.max()+y.min())/2, (z.max()+z.min())/2)):
        axis(center - Xb, center + Xb)

    plt.show()
    return fig, ax

plot_3d_from_df(pd.read_csv('../matches/HYDRA-0057_match.csv'))
```
